```python
import uproot
import matplotlib.pyplot as plt
%matplotlib inline



```

```python
f = uproot.open("/home/rice/Projects/DUNE_Opticks/aaron_test/debug_test/")

f.keys()

edep=f["optical/SimEnergyDeposit"]
```

```python
evtId=edep["evtID"].array().to_numpy()
x=edep["x"].array().to_numpy()
y=edep["y"].array().to_numpy()
z=edep["z"].array().to_numpy()
nphot=edep["nphot"].array().to_numpy()
```

```python
EventID=4
eventMask=evtId==EventID
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
fig, ax = plt.subplots()
ax.set_title(" A Particle Track")
h=ax.hist2d(z[eventMask],y[eventMask],weights=nphot[eventMask],bins=120, vmin=0, vmax=40000)

ax.xaxis.set_major_locator(ticker.MultipleLocator(200))  # 200‑unit spacing
# Add colorbar (belongs to the figure)
cbar = fig.colorbar((h[3]), ax=ax)
cbar.formatter.set_powerlimits((0, 0))
cbar.set_label(r"Simulated Photons")

# Axis labels
ax.set_xlabel("z (cm)")
ax.set_ylabel("y (cm)")
```
